## **Internal Intensity Training Data**  

### **Context**
This internal dataset consists of one individual's (Nono's) training data over seven years of triathlon training. We analysed Nono’s personal training data over two distinct periods: high-intensity and low-intensity training. The goal was to evaluate changes in resting heart rate (RHR) and determine the effects of different training intensities on cardiovascular fitness.  

### **Aim** 
Our objective was to investigate whether high-intensity training leads to a greater reduction in RHR compared to low-intensity training. Additionally, we examined the statistical significance of these changes.  

### **Process**

#### **High-Intensity Training Period** 
- RHR was measured 4 weeks before and 4 weeks after starting high-intensity running.  
- A correlation test was conducted to assess whether starting with a higher RHR leads to a larger percentage decrease in RHR.  
- Mann-Whitney U-Test was used to determine if there was a statistically significant difference in RHR before and after high-intensity training.  

#### **Low-Intensity Training Period** 
- RHR was measured 4 weeks before and 4 weeks after a continuous low-intensity training phase.  
- Mann-Whitney U-Test was applied to compare pre- and post-training RHR values.  

### **Observations** 

#### **High-Intensity Training Period**  
- Observed RHR decrease: 1.9 bpm.  
- Correlation Analysis: A correlation was found between initial RHR and the percentage decrease in RHR.  
- Significance Testing: The Mann-Whitney U-Test confirmed a statistically significant difference in RHR before and after high-intensity training.  

#### **Low-Intensity Training Period**  
- Observed RHR increase: 0.6 bpm.  
- Significance Testing:  
  - U = 412.0, p-value = 0.6457  
  - Interpretation: The high p-value (> 0.05) indicates no significant difference in RHR before and after low-intensity training, suggesting RHR returned to a similar level post-training. So we can assume this is likely by chance.

### **Conclusion & Further Insights**  
- Any form of training can contribute to lowering RHR and improving cardiovascular fitness.  
- Both trial data and Nono’s data suggest that high-intensity training has a stronger impact on reducing RHR.  
- Endurance training programs should incorporate high-intensity sessions for optimal cardiovascular benefits.  
- Further analysis could explore long-term effects, individual variability, and the influence of age or gender on training adaptations.

Checking Corelation coef. to test whether starting with a higher resting HR leads to a larger % decrease in resting HR

```python
import pandas as pd
from scipy.stats import pearsonr

# Ensure 'date' column is a datetime object
df_phys_hi["date"] = pd.to_datetime(df_phys_hi["date"])

# Define time periods
start_date_b = '2021-12-16'
end_date_b = '2022-01-13'
start_date_a = '2022-04-10'
end_date_a = '2022-05-08'

# Filter data for before and after HIIT periods
df_before = df_phys_hi[(df_phys_hi['date'] >= start_date_b) & (df_phys_hi['date'] <= end_date_b)]
df_after = df_phys_hi[(df_phys_hi['date'] >= start_date_a) & (df_phys_hi['date'] <= end_date_a)]

# Calculate mean RHR before and after HIIT
avg_rhr_before = df_before["rhr"].mean()
avg_rhr_after = df_after["rhr"].mean()

# Compute % change in RHR
hr_change_percent = ((avg_rhr_before - avg_rhr_after) / avg_rhr_before) * 100

# Correlation test (since we don’t have participant-level data, this is for illustration)
corr_coef, p_value = pearsonr(df_before["rhr"], df_after["rhr"])

# Print results
print(f"Avg RHR Before HIIT: {avg_rhr_before:.2f}")
print(f"Avg RHR After HIIT: {avg_rhr_after:.2f}")
print(f"Percentage Change in RHR: {hr_change_percent:.2f}%")
print(f"Correlation between Pre-RHR and Post-RHR Change: {corr_coef:.4f} (p={p_value:.4f})")
```

Mann-Whitney U test

```python
import scipy.stats as stats

# Define the three groups using the date filters:
rhr_before = df_phys_hi[(df_phys_hi['date'] >= '2021-12-16') & (df_phys_hi['date'] <= '2022-01-13')]['rhr']
rhr_during = df_phys_hi[(df_phys_hi['date'] >= '2022-01-13') & (df_phys_hi['date'] <= '2022-04-10')]['rhr']
rhr_after  = df_phys_hi[(df_phys_hi['date'] >= '2022-04-10') & (df_phys_hi['date'] <= '2022-05-08')]['rhr']

# Perform pairwise Mann-Whitney U tests (non-parametric test for two independent samples):

# Before vs. During
u_stat1, p_val1 = stats.mannwhitneyu(rhr_before, rhr_during, alternative='two-sided')

# Before vs. After
u_stat2, p_val2 = stats.mannwhitneyu(rhr_before, rhr_after, alternative='two-sided')

# During vs. After
u_stat3, p_val3 = stats.mannwhitneyu(rhr_during, rhr_after, alternative='two-sided')

print("Mann-Whitney U Test Results:")
print("Before vs. During: U =", u_stat1, ", p-value =", p_val1)
print("Before vs. After:  U =", u_stat2, ", p-value =", p_val2)
print("During vs. After:  U =", u_stat3, ", p-value =", p_val3)
```

Before vs. During:

p-value = 0.228
This p-value is above the conventional threshold of 0.05, which suggests that there is no statistically significant difference in resting heart rate between the period before training and the period during training.- (perhaps we dont need this anyway)
#
Before vs. After:

p-value = 0.00416
This p-value is below 0.05, indicating a statistically significant difference in resting heart rate between the period before training and the period after training.
#
During vs. After:

p-value = 0.00919
Similarly, this p-value is below 0.05, suggesting a statistically significant difference in resting heart rate between the period during training and the period after training.

Mann-whitney test

```python
import scipy.stats as stats

# Define the three groups using the given date filters:
rhr_before = df_phys_hi[(df_phys_hi['date'] >= '2023-01-28') & (df_phys_hi['date'] <= '2023-02-28')]['rhr']
rhr_during = df_phys_hi[(df_phys_hi['date'] >= '2023-03-01') & (df_phys_hi['date'] <= '2023-05-30')]['rhr']
rhr_after  = df_phys_hi[(df_phys_hi['date'] >= '2023-06-01') & (df_phys_hi['date'] <= '2023-06-30')]['rhr']

# Perform pairwise Mann-Whitney U tests (non-parametric test for two independent samples):

# Before vs. During
u_stat1, p_val1 = stats.mannwhitneyu(rhr_before, rhr_during, alternative='two-sided')

# Before vs. After
u_stat2, p_val2 = stats.mannwhitneyu(rhr_before, rhr_after, alternative='two-sided')

# During vs. After
u_stat3, p_val3 = stats.mannwhitneyu(rhr_during, rhr_after, alternative='two-sided')

# Print results
print("Mann-Whitney U Test Results:")
print("Before vs. During: U =", u_stat1, ", p-value =", p_val1)
print("Before vs. After:  U =", u_stat2, ", p-value =", p_val2)
print("During vs. After:  U =", u_stat3, ", p-value =", p_val3)
```
